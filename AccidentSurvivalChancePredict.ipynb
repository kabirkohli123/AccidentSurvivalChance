```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Accident_train.csv")
df
```

```python
for i in range(17):
    nullCheck = df.iloc[:,i].isnull().sum()
    print("The number of null value in column ",i," are: ",nullCheck)
```

```python
df = df.dropna()
df
```

```python
for i in range(17):
    nullCheck = df.iloc[:i].isnull().sum()
    print("The number of null value in column ",i," are: ",nullCheck)
```

```python
df.Hour_of_Collision.unique()
```

```python
df = df.reset_index(drop=True)
df
```

```python
dummies1 = pd.get_dummies(df['Weekday_of_Collision'])
dummies1
```

```python
dummies2 = pd.get_dummies(df['Policing_Area'])
dummies2
```

```python
df.drop(['Policing_Area','Weekday_of_Collision'],axis='columns',inplace=True)
df = pd.concat([df,dummies1,dummies2],axis='columns')
df
```

```python
Y = df['Collision_Severity']
Y.unique()
```

```python
X = df.drop(['Collision_Severity','Collision_Ref_No'],axis='columns')
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,train_size=0.8,random_state=7)
X_test
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,Y_train)
Y_predict = lr.predict(X_test)
Y_predict[:10]
```

```python
print("The first 10 test value of Y is ")
print(np.array(Y_test[:10]))
print("The first 10 predicted value of Y is ")
print(Y_predict[:10])
print("The Accuracy of Logistic Regression is ")
print(lr.score(X_test,Y_test))
```

```python
from sklearn import metrics
matrix = metrics.confusion_matrix(Y_predict,Y_test)
sns.heatmap(matrix,annot=True,cmap='Blues',fmt='0.1f')
plt.xlabel("Predicted Values")
plt.ylabel("Test Value")
plt.title("Confusion Matrix for Logistic Regression")
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_train,Y_train)
Y_predict = nb.predict(X_test)
print("The first 10 test value of Y is ")
print(np.array(Y_test[:10]))
print("The first 10 predicted value of Y is ")
print(Y_predict[:10])
print("The accuracy of Naive Bayes is ")
print(nb.score(X_test,Y_test))
```

```python
matrix = metrics.confusion_matrix(Y_predict,Y_test)
sns.heatmap(matrix,annot=True,cmap='Purples',fmt='0.1f')
plt.xlabel("Predicted Values")
plt.ylabel("Test Value")
plt.title("Confusion Matrix for Naive Bayes")
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,Y_train)
Y_predict = knn.predict(X_test)
print("The first 10 test value of Y is ")
print(np.array(Y_test[:10]))
print("The first 10 predicted value of Y is ")
print(Y_predict[:10])
print("The accuracy of Naive Bayes is ")
print(knn.score(X_test,Y_test))
```

```python
matrix = metrics.confusion_matrix(Y_predict,Y_test)
sns.heatmap(matrix,annot=True,cmap='Purples',fmt='0.1f')
plt.xlabel("Predicted Values")
plt.ylabel("Test Value")
plt.title("Confusion Matrix for KNN")
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train,Y_train)
Y_predict = dt.predict(X_test)
print("The first 10 test value of Y is ")
print(np.array(Y_test[:10]))
print("The first 10 predicted value of Y is ")
print(Y_predict[:10])
print("The accuracy of Naive Bayes is ")
print(dt.score(X_test,Y_test))
```

```python
matrix = metrics.confusion_matrix(Y_predict,Y_test)
sns.heatmap(matrix,annot=True,cmap='Purples',fmt='0.1f')
plt.xlabel("Predicted Values")
plt.ylabel("Test Value")
plt.title("Confusion Matrix for Decision Tree Classifier")
```

```python
from sklearn.model_selection import cross_val_score
lr.fit(X,Y)
print("The cross val score of Logistic Regression is :")
print(cross_val_score(lr,X,Y,cv=10))
print("The average cross val score of Logistic Regression is :")
print(np.average(cross_val_score(lr,X,Y,cv=10)))
```

```python
nb.fit(X,Y)
print("The cross val score of Naive Bayes is :")
print(cross_val_score(nb,X,Y,cv=10))
print("The average cross val score of Naive Bayes is :")
print(np.average(cross_val_score(nb,X,Y,cv=10)))
```

```python
knn.fit(X,Y)
print("The cross val score of KNN is :")
print(cross_val_score(knn,X,Y,cv=10))
print("The average cross val score of KNN is :")
print(np.average(cross_val_score(knn,X,Y,cv=10)))
```

```python
dt.fit(X,Y)
print("The cross val score of Decision Tree Classifier is :")
print(cross_val_score(dt,X,Y,cv=10))
print("The average cross val score of Decision Tree Classifier is :")
print(np.average(cross_val_score(dt,X,Y,cv=10)))
```
